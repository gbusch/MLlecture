# KMeans Clustering #

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

The make_blobs command can be imported from scikit-learn and creates a mock data set to train with. The y array contains labels, but since we are practising an unsupervised ML algorithm, we will ignore these labels and only work with the features.

```python
X, y= make_blobs(random_state=1)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

The KMeans algorithm is included in scikit-learn and can be called as a class. Here, we specify the number of clusters to be fitted. From the inspection of the data, we chose 3, but we could also try other values here.

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

We can now print the fitted labels for all data points, as well as the cluser centres and plot them in a diagram. We see that the algorithm found the clusters automatically without any additional label information.

```python
print("Cluster memberships:",kmeans.labels_)
```

```python
print("Cluster centres:",kmeans.cluster_centers_)
```

```python
plt.close()
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='^',c='black')
plt.show()
```
